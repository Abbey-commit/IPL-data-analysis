```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load Dataset

```python
data = pd.read_csv("IPL Ball-by-Ball 2008-2020.csv")
```

```python
#Drop the field feature
data.drop(['fielder', 'player_dismissed', 'dismissal_kind', 'extras_type'], axis=1, inplace=True)
```

```python
temp = pd.DataFrame({'Bowler':data['bowler']})
count_bowls = temp.value_counts()
print(count_bowls)
```

```python
labels = [X[0] for X in count_bowls.keys()]

bar, ax = plt.subplots(figsize=(20,12))
ax = plt.pie(x = count_bowls, labels=labels)
plt.title("Most Wins in IPL", fontsize=15)
```

```python
teams = data['bowling_team'].unique()
decision_making = pd.DataFrame([], columns= ['Bowling Team', 'Decision', 'Times'])

for id, element in enumerate(teams):
    temp_bowl = data[(data['bowling_team'] == 'Royal Challengers Bangalore') & (data['bowler'] == 'AA Noffke')]
    temp_bat = data[(data['batting_team'] == 'Kolkata Knighter Rider') & (data['batsman'] == 'RT Ponting')]
    
    #appending to the decision making
    decision_making = decision_making.append({'Bowling Team': element,
                                      'Decision': 'bowl', 'Times':temp_bowl['bowling_team'].count()}, ignore_index=True)
    decision_making = decision_making.append({'Batting Team': element,
                                             'Decision':'bat', 'Times':temp_bat['batting_team'].count()}, ignore_index=True)
```

```python
decision_making
```

```python
sns.catplot(x='Bowling Team', y='Times', hue='Decision', data=decision_making, kind='bar', height=5, aspect=2)

plt.xticks(rotation=90)
plt.title("Bowling Decision of Teams")
plt.xlabel("IPL Teams")
plt.ylabel(" Bowl Decision ")
```
